## 由以下七題選擇三題完成

### 1. Invert

```python
# image invert #
# you must plot both the original image and the modefied image as the result.
# -------- To do ------------- #
image = cv2.imread('image/duck.jpg')
image = image[:,:,::-1]
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.title('original')
plt.imshow(image)


invert = [255, 255, 255] - image
plt.subplot(122)
plt.title('invert')
plt.imshow(invert)


plt.show()
```

### 2. Darken

```python
# image datken #
# you must plot both the original image and the modefied image as the result.
# -------- To do ------------- #
image = cv2.imread('image/duck.jpg')
image = image[:,:,::-1]
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.title('original')
plt.imshow(image)


darken = image

for i in range(len(darken)):
  for j in range(len(darken[i])):
    for k in range(3):
      if darken[i][j][k] - 128 >= 0:
        darken[i][j][k] -= 128
      else:
        darken[i][j][k] = 0

plt.subplot(122)
plt.title('darken')
plt.imshow(darken)


plt.show()
```

### 3. Raise Lighten

```python
# image lighten #
# you must plot both the original image and the modefied image as the result.
# -------- To do ------------- #
image = cv2.imread('image/duck.jpg')
image = image[:,:,::-1]
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.title('original')
plt.imshow(image)


lighten = image

for i in range(len(lighten)):
  for j in range(len(lighten[i])):
    for k in range(3):
      if lighten[i][j][k] + 128 <= 255:
        lighten[i][j][k] += 128
      else:
        lighten[i][j][k] = 255

plt.subplot(122)
plt.title('lighten')
plt.imshow(lighten)


plt.show()
```

### 4. Lower Contrast

```python
# image Lower Contrast #
# you must plot both the original image and the modefied image as the result.
# -------- To do ------------- #
image = cv2.imread('image/duck.jpg')
image = image[:,:,::-1]
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.title('original')
plt.imshow(image)


lower_contrast = image
for i in range(len(lower_contrast)):
  for j in range(len(lower_contrast[i])):
    for k in range(3):
      lower_contrast[i][j][k] /= 2

plt.subplot(122)
plt.title('lower contrast')
plt.imshow(lower_contrast)


plt.show()
```

### 5. Raise Contrast

```python
# image raise contrast #
# you must plot both the original image and the modefied image as the result.
# -------- To do ------------- #
image = cv2.imread('image/duck.jpg')
image = image[:,:,::-1]
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.title('original')
plt.imshow(image)


raise_contrast = image
for i in range(len(raise_contrast)):
  for j in range(len(raise_contrast[i])):
    for k in range(3):
      if raise_contrast[i][j][k] *2 <=255:
        raise_contrast[i][j][k] *= 2
      else:
        raise_contrast[i][j][k] = 255
      

plt.subplot(122)
plt.title('raise contrast')
plt.imshow(raise_contrast)


plt.show()
```

### 6. Non-linear Lower Contrast

```python
# Non-linear lower contrast #
# you must plot both the original image and the modefied image as the result.
# -------- To do ------------- #
image = cv2.imread('image/duck.jpg')
image = image[:,:,::-1]
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.title('original')
plt.imshow(image)


non_linear_lower = image
for i in range(len(non_linear_lower)):
  for j in range(len(non_linear_lower[i])):
    for k in range(3):
      if (non_linear_lower[i][j][k] / 255) ** (1 / 3) *255 >= 0:
        non_linear_lower[i][j][k] = (non_linear_lower[i][j][k] / 255) ** (1 / 3) *255
      else:
        non_linear_lower[i][j][k] = 0
      

plt.subplot(122)
plt.title('non-linear lower contrast')
plt.imshow(non_linear_lower)


plt.show()
```

### 7. Non-linear Raise Contrast

```python
# Non-linear raise contrast #
# you must plot both the original image and the modefied image as the result.
# -------- To do ------------- #
image = cv2.imread('image/duck.jpg')
image = image[:,:,::-1]
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.title('original')
plt.imshow(image)


non_linear_raise = image
for i in range(len(non_linear_raise)):
  for j in range(len(non_linear_raise[i])):
    for k in range(3):
      if (non_linear_raise[i][j][k] / 255) ** 2 *255 <= 255:
        non_linear_raise[i][j][k] = (non_linear_raise[i][j][k] / 255) ** 2 *255
      else:
        non_linear_raise[i][j][k] = 255
      

plt.subplot(122)
plt.title('non-linear raise contrast')
plt.imshow(non_linear_raise)


plt.show()
```
